```python
import pandas as pd
import altair as alt
```

```python
cpi = pd.read_csv('./USACPIALLMINMEI.csv', names = ['date', 'cpi'],header = 0)
cpi
```

```python
cpi.date = pd.to_datetime(cpi.date, format = "%Y-%m-%d")
cpi
```

```python
alt.Chart(cpi).mark_line().encode(
    x = 'date', 
    y = 'cpi:Q'
)
```
