```python
us = cases.loc[cases.Country == 'United States of America']
us.Date_reported = pd.to_datetime(us.Date_reported, format = '%Y-%m-%d')
```

```python
us
```

```python
import altair as alt
```

```python
alt.Chart(us).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'New_cases'
)
```

```python
alt.Chart(us).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'New_deaths'
)
```

```python
alt.Chart(us).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'Cumulative_cases'
)
```

```python
alt.Chart(us).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'Cumulative_deaths'
)
```

```python
alt.Chart(
    us.loc[us.Date_reported > pd.Timestamp('07-15-2021')]
).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'New_cases'
)
```

```python
alt.Chart(
    us.loc[us.Date_reported > pd.Timestamp('07-15-2021')]
).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'Cumulative_cases'
)
```

```python
us_delta = us.loc[us.Date_reported > pd.Timestamp('07-15-2021')].copy()
```

```python

us_delta.Cumulative_cases \
- us_delta.Cumulative_cases.min()
```

```python
us_delta['new_cumulative_cases'] = us_delta.Cumulative_cases - us_delta.Cumulative_cases.min()
```

```python
alt.Chart(
    us_delta
).mark_line().encode(
    x = 'Date_reported:T', 
    y = 'new_cumulative_cases'
)
```

```python
us_delta.to_csv('./us_july_15-present.csv')
```
